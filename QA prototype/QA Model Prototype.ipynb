```python
# {model training} 

#model intialisation from one of the pre trained models provided by the sentence-transformers library. 
model = SentenceTransformer('stsb-roberta-base')

# X is the collection of all the question vectors
X = model.encode(questions)

#Y is the collection of all the answer vectors
Y = model.encode(answers)

#R is the trainable transformation matrix
R = np.random.random((768,768))
XR = np.dot(X, R)

#learning rate alpha
alpha = 0.01
```

```python
print(np.shape(Y))
```

```python
#training loop
for i in range(300):
    loss = XR - Y
    
    #Frobenius loss for simpler calculation
    loss_Frobenius = np.sum(np.square(loss))
    
    #gradiet of the loss function w.r.t R
    g = 2*(np.dot(np.transpose(X), loss))/len(questions)
    
    #updating R
    R = R - alpha*g
    
print(loss_Frobenius)
```

```python
#sample query which is already present in the training set
query = ["Which country are you from"]

#query encoded into a vector
query_embed = model.encode(query)

#query vector mapped into the answer vector space by the transformation matrix R
ans_vec = np.dot(query_embed, R)

#array containing the cosine scores of the answer vector with the answer vectors in the training set to measure similarity 
# of the answer vector with the training set answer vectors
cosine_scores = sklearn.metrics.pairwise.cosine_similarity(ans_vec, Y)
print(cosine_scores)
```

```python
print(dict['Which country are you from'])
```

```python
score_list=[]
for i in cosine_scores:
    for j in i:
        score_list.append(j)
index = score_list.index(max(score_list))

#prints the answer that has the highest cosine score w.r.t to the answer vector
print(answers[index])
```

```python
#measures the euclidean distance of the answer vector with the answer vector in the training set, to implement nearest
# neighbours

distances = []
for i in Y:
    dist = np.linalg.norm(ans_vec - i)
    distances.append(dist)
print(distances)
print(answers[distances.index(min(distances))])
```
